# SDSS DR7

```python
lss_data = np.loadtxt('../data/SDSS/dr72_bright_34_lss/lss.dr72bright34.dat')
photo_data = np.loadtxt('../data/SDSS/dr72_bright_34_lss/photoinfo.dr72bright34.dat')
```

```python
c = 299792.458
z_obs = lss_data[:,5]/c
ra = lss_data[:,3]
dec = lss_data[:,4]
m_r = photo_data[:,3]

ii = (z_obs<=0.114) & (m_r<=-20.0)
ii_radec = (dec>0) & (dec<50) & (ra>130) & (ra<230)
```

```python
plt.scatter(ra[ii&ii_radec], dec[ii&ii_radec], s=0.1)
```

```python
phi   = np.deg2rad(360 - ra[ii&ii_radec])
theta = np.deg2rad(90 - dec[ii&ii_radec])
r = np.array(cosmo.comoving_distance(z_obs[ii&ii_radec]))
mag_r = m_r[ii&ii_radec]

x = r * np.cos(phi) * np.sin(theta)
y = r * np.sin(phi) * np.sin(theta)
z = r * np.cos(theta)

data = np.array([x, y, z, mag_r])
print(len(x))
```

```python
plt.figure(figsize=(10,10))
ii_slice = (z>50) & (z<80)
plt.scatter(y[ii_slice], -x[ii_slice], c=mag_r[ii_slice], alpha=0.1)
_ = plt.axis("equal")
```

```python
output_dir = "../data/SDSS/pos/"
data_filename = "mag_r_limited_xyz.dat"
np.savetxt(os.path.join(output_dir, data_filename), data.T, 
           header="x[Mpc] y[Mpc] z[Mpc] M_r")
```

```python
def make_random_data(phi, theta, r):
    min_phi = np.min(phi)
    max_phi = np.max(phi)
    min_costheta = np.min(np.cos(theta))
    max_costheta = np.max(np.cos(theta))
    n = len(phi)
    random_phi = np.random.random(n)*(max_phi - min_phi) + min_phi
    random_theta = np.arccos(np.random.random(n)*(max_costheta - min_costheta) + min_costheta)
    random_r = (np.random.random(n)**(1.0/3.0)) * np.max(r) 

    random_x = random_r * np.cos(random_phi) * np.sin(random_theta)
    random_y = random_r * np.sin(random_phi) * np.sin(random_theta)
    random_z = random_r * np.cos(random_theta)
    return np.array([random_x, random_y, random_z])
```

```python
n_random = 100
for i in range(n_random):
    random_data  = make_random_data(phi, theta, r)
    
    random_filename = "random_{:03}_".format(i) + data_filename
    print(random_filename)
    np.savetxt(os.path.join(output_dir, random_filename), random_data.T, 
           header="x[Mpc] y[Mpc] z[Mpc]")
print(n_points)
```
