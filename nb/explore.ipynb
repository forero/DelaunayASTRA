```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('../data/fof_massive_xyz_mass_200Mpc512.txt')
data_connections = np.loadtxt('../data/fof_massive_xyz_mass_200Mpc512_n_connections.txt')
data_web = np.loadtxt('../data/web_fof_massive_xyz_mass_200Mpc512.txt')
rand = np.loadtxt('../data/001_random_fof_massive_xyz_mass_200Mpc512.txt')
rand_connections = np.loadtxt('../data/001_random_fof_massive_xyz_mass_200Mpc512_n_connections.txt')
rand_web = np.loadtxt('../data/web_001_random_fof_massive_xyz_mass_200Mpc512.txt')
```

```python
r_data = (data_connections[:,0]-data_connections[:,1])/(data_connections[:,0]+data_connections[:,1])
r_rand = (rand_connections[:,0]-rand_connections[:,1])/(rand_connections[:,0]+rand_connections[:,1])

plt.hist(r_data, bins=27)
plt.hist(r_rand, alpha=0.5, bins=27)
```

```python
np.count_nonzero(r_rand<-0.50)/len(r_rand)
```

```python
ii = (data[:,2]>100) & (data[:,2]<120) 
plt.scatter(data[ii,0], data[ii,1], c=r_data[ii], s=10, cmap='seismic')
plt.colorbar()
```

```python
ii = (data[:,2]>100) & (data[:,2]<120) 
plt.scatter(data[ii,0], data[ii,1], c=data_web[ii], s=10, cmap='gnuplot')
plt.colorbar()
```

```python
ii = (rand[:,2]>100) & (rand[:,2]<120)
plt.scatter(rand[ii,0], rand[ii,1], c=r_rand[ii], s=10,cmap='seismic')
plt.colorbar()
```

```python
ii = (rand[:,2]>100) & (rand[:,2]<120)
plt.scatter(rand[ii,0], rand[ii,1], c=rand_web[ii], s=10, cmap='gnuplot')
plt.colorbar()
```
